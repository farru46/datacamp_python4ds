# Pattern Matching

```python
# Import the regular expression module
import re
```

```python
# Compile the pattern: prog
prog = re.compile('\d{3}-\d{3}-\d{4}')
```

```python
# See if the pattern matches
result = prog.match('123-456-7890')
print(bool(result))
```

```python
# See if the pattern matches
result2 = prog.match('1123-456-7890')
print(bool(result2))
```

```python
# Recoding the sex column to recode column in tips data
```

```python
from numpy import NaN
```

```python
def gender_recode(gender):
    if gender.lower() == 'male':
        return 1
    elif gender.lower() == 'female':
        return 0
    else:
        return NaN
```

```python
tips['recoded gender'] = tips['sex'].apply(gender_recode)
```
