# Downsampling and Upsampling

```python
weather_data_clean.resample('D').mean()
```

```python
weather_data_clean.head()
```

```python
weather_data_clean['sky_condition'].unique().tolist()
```

```python
sunny = weather_data_clean[weather_data_clean['sky_condition'] == 'CLR']
```

```python
sunny.head()
```

```python
sunny.resample('D').max()
```

```python
is_sky_overcast = weather_data_clean['sky_condition'].str.contains('OVC')
```

```python
is_sky_overcast
```

```python
overcast = weather_data_clean[is_sky_overcast]
```

```python
overcast.head()
```

```python
overcast_daily_max = overcast.resample('D').max()
```

```python
overcast_daily_max.head()
```

```python
is_sky_clear = weather_data_clean['sky_condition']=='CLR'
sunny = weather_data_clean.loc[is_sky_clear]
sunny_daily_max = sunny.resample('D').max()
is_sky_overcast = weather_data_clean['sky_condition'].str.contains('OVC')
overcast = weather_data_clean.loc[is_sky_overcast]
overcast_daily_max = overcast.resample('D').max()
```

```python
sunny_daily_max.head()
```

```python
# Calculate the mean of sunny_daily_max
sunny_daily_max_mean = sunny_daily_max.mean()

# Calculate the mean of overcast_daily_max
overcast_daily_max_mean = overcast_daily_max.mean()

# Print the difference (sunny minus overcast)
print(sunny_daily_max_mean - overcast_daily_max_mean )
```
