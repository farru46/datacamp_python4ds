# Higher Order Regressions

```python
# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto['weight'], auto['mpg'], label='data', color='red', marker='o')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='blue', scatter=None, label='First Order')

# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='weight', y='mpg', data=auto, color='green', scatter=None, label='Second Order',order=2)

# Add a legend and display the plot
plt.legend(loc='upper right')
plt.show()
```

# Visualizing Univariate Distributions

* Strip Plots 
* Swarm Plots
* Violin Plots

```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
auto=pd.read_csv("auto-mpg.csv", na_values='?')
auto
```

### Strip Plot

```python
# Make a strip plot of 'hp' grouped by 'cyl'
plt.subplot(2,1,1)
sns.stripplot(x='cylinders', y='horsepower', data=auto)

# Make the strip plot again using jitter and a smaller point size
plt.subplot(2,1,2)
sns.stripplot(x='cylinders', y='horsepower', data=auto, jitter=True, size =3)

# Display the plot
plt.show()
```

### Swarm Plot

```python
sns.swarmplot(x='cylinders',y='horsepower',data=auto)
```

```python
sns.swarmplot(x='cylinders',y='horsepower',data=auto,hue='origin')
```

```python
sns.swarmplot(y='cylinders',x='horsepower',data=auto,hue='origin',orient='h')
```

### Violin plots

```python
sns.violinplot(x='cylinders',y='horsepower',data=auto,inner=None,color='lightgray')
sns.stripplot(x='cylinders',y='horsepower',data=auto,jitter=True)
```
