```python
data=pd.read_csv("../../datasets/stock_data", header=None,names=['Month', 'A', 'B', 'C'])
```

```python
data
```

```python
fig, ax = plt.subplots()
ax.plot(data['Month'], data.A)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data['Month'], data['A'], linestyle='--', marker="o")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data['Month'], data['A'], linestyle='None', marker="o")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data['Month'], data['C'], color='r')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data['Month'], data['B'], marker='o', color='blue', linestyle="--")
ax.set_xlabel('Month')
ax.set_ylabel('Price')
```

```python
fig, ax=plt.subplots()
ax.plot(data['Month'],data['A'],marker='o')
ax.plot(data['Month'],data['B'], marker='^', linestyle='--')
plt.show()
```

```python
arr1=np.random.rand(20,)
```

```python
arr1
```

```python
arr2 = np.random.rand(20,)
```

```python
index=np.arange(20)
```

```python
index
```

```python
plt.plot(index, arr1, color = 'red')
plt.plot(index, arr2, color='blue' )
```

```python
fig, ax=plt.subplots()
ax.plot(index,arr1,marker='o')
ax.plot(index,arr2,marker='o', linestyle='--')
plt.show()
```

```python
# Using Axes
```

```python
plt.axes([0.05, 0.05, .425, 0.9 ])
plt.plot(index ,arr1, color='red')
plt.xlabel('Label1')
plt.axes([.7, 0.05, .425,0.9])
plt.plot(index, arr2, color='blue')
```

```python
plt.subplot(2,1, 2)
```

```python
plt.subplot(1,2,1)
plt.plot(index, arr1,color='red')
plt.title('First Plot')
plt.subplot(1,2,2)
plt.plot(index, arr2,color='blue')
```
